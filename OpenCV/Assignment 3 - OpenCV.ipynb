```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Edges in previous image

### Using Canny

```python
img = cv2.imread('/home/pranavkdas/Documents/nemo.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img,100,200)
plt.figure(figsize=(8,8))
plt.imshow(edges)
plt.show()
```

### Without Canny

```python
img = cv2.imread('/home/pranavkdas/Documents/nemo.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (21,21), 0, 0)
img_blend = cv2.divide(img_gray, img_blur, scale=256)
plt.figure(figsize=(8,8))
plt.imshow(img_blend,'gray')
plt.show()
```

### B&W

```python
img = cv2.imread('/home/pranavkdas/Documents/nemo.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(grayscale,(5,5),0)
bw = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
laplacian = cv2.Laplacian(bw,cv2.CV_8U)
plt.figure(figsize=(8,8))
plt.imshow(bw,'gray')
plt.show()
```

# Pencil sketch on Video

### Using Canny

```python
capture = cv2.VideoCapture(0) 
i=200 
  
while(i): 
    ret, frame = capture.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(frame,100,200) 
    cv2.imshow('frame',edges) 
    cv2.waitKey(5)
    i=i-1
    
capture.release() 
cv2.destroyAllWindows()  
```

### Without Canny

```python
capture = cv2.VideoCapture(0) 
i=200 
  
while(i): 
    ret, img = capture.read()
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_blur = cv2.GaussianBlur(img_gray, (21,21), 0, 0)
    img_blend = cv2.divide(img_gray, img_blur, scale=256)
    cv2.imshow('frame',img_blend) 
    cv2.waitKey(5)
    i=i-1
    
capture.release() 
cv2.destroyAllWindows()
```
